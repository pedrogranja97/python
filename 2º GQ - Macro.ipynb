### **1. Explique o conceito de NAIRU.**

A sigla significa *non-accelerating inflation rate of unemployment*, ou **taxa de desemprego não aceleradora da inflação.**

Trata-se de uma teoria desenvolvida por Edmund Phelps, que fala da menor taxa de desemprego que não aumenta a inflação. Se diz que se a taxa de desemprego está **no nível da NAIRU, a inflação é constante** - obviamente, a inflação não tem apenas uma única causa, então, como toda teoria econômica, ela assume ceteris paribus. Sendo assim, se o desemprego subir, ocorre deflação, caso contrário, inflação. \
A teoria está **intimamente conectada à ideia de taxa natural de desemprego**. Banco centrais utilizam a NAIRU, ou, melhor dizendo, sua estimativa de NAIRU, para tomar decisões referentes à taxa de juros no país. \
**Não existe uma fórmula definida para a NAIRU**. Normalmente são utilizados modelos estatísticos para estimá-la, com a maioria dos bancos centrais estimando um valor em volta de 6%. Nesse ponto, um **aumento na demanda agregada não irá diminuir o desemprego estrutural e/ou friccional**, apenas aumentar os salários nominais, aumentando a inflação. Afinal, o aumento na oferta de empregos combinado à pouca capacidade de absorção das vagas aumentaria a barganha dos empregados. \
Os salários mais altos, sem aumento de produção, **obrigaria as empresas a diminuírem seu quadro de funcionários**. Então, estaríamos de volta a uma taxa natural de desemprego, só que com aceleração da inflação.

```python
# Dados

x = np.array([5, 6, 7, 8])
y = np.array([14, 6, 3, 2])
z = np.array([6, 7, 8, 9])
k = np.array([10.5, 7, 6, 4])
x1, y1= [6, 6], [20, 0]

X_Y_Spline = make_interp_spline(x, y)
Z_K_Spline = make_interp_spline(z, k)
 
X_ = np.linspace(x.min(), x.max(), 15)
Z_ = np.linspace(x.min(), x.max(), 15)
Y_ = X_Y_Spline(X_)
K_ = Z_K_Spline(Z_)
 
# Plotando gráfico
plt.title('Modelo da NAIRU', fontname = 'Franklin Gothic Book', fontsize = 18)
plt.plot(X_, Y_, label = 'Curva de Phillips no curto prazo', color = 'black')
plt.plot(Z_, K_, label = 'Nova Curva de Phillips no curto prazo', color = 'orange')
plt.plot(x1, y1, linestyle = 'dashed', label = 'NAIRU', color = 'green')
plt.legend(fontsize = 10)
plt.xlabel("Desemprego (%)", fontname = 'Franklin Gothic Book', fontsize = 14)
plt.ylabel("Inflação (%)", fontname = 'Franklin Gothic Book', fontsize = 14)
plt.ylim(0, 20)

plt.show()
```

Como podemos perceber no gráfico de exemplo, a taxa de desemprego continua nos 6%, mas **com novo patamar inflacionário.** Mesmo com todo esforço, é virtualmente impossível manter a taxa de desemprego em 0% no longo prazo, pois sempre haverá desemprego friccional e desencontro de informações. Para diminuir a taxa natural de desemprego, **é necessário aumento de produtividade**, mais comumente encontrado através do avanço da tecnologia. Ao longo do tempo e com o avanço da produtividade, se percebe que a taxa de desemprego natural vem caindo, movendo a NAIRU para a esquerda (no gráfico). **A NAIRU também é chamada de Curva de Phillips no longo prazo.**

### **2. Por que, segundo Solow, as economias que não apresentam progresso tecnológico, não crescem no longo prazo?**

Robert Solow, ganhador do prêmio Nobel de Economia em 1987, é um economista americano que desenvolveu o Modelo de Solow-Swan. Que busca explicar as razões porque alguns países são mais ricos que outros. O modelo de Solow, que busca explicar como variáveis como capital, força de trabalho e produtividade, influenciam o crescimento econômico ao longo do tempo. <br />
Sem crescimento tecnológico, o crescimento per capita vai cessar no momento em que começar o retorno decrescente de capital, o que é chamado de estado estacionário (nível de estoque de capital em que o investimento é igual à depreciação de capital). O aumento da poupança também contribui para o aumento de capital por trabalhador, mas tem efeito limitado, com a economia voltando ao equilíbrio de longo prazo. <br />
Porém, quando adicionada a variável de tecnologia, que é considerada exógena, se percebe um constante avanço de produtividade, o que causa um crescimento sustentável no longo prazo. Isso acontece porque a tecnologia permite aumento em diversas variáveis que afetam o crescimento econômico, entre elas: população, capital e produtividade.
