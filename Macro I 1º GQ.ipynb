### 1ª questão

Mostre através do modelo IS/LM como um aumento no risco em uma
economia afeta o nível de atividade.

```python
### importando pacotes necessários

import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('fivethirtyeight')
```

#### Resposta

```python
### plotando gráfico IS-LM

x1, y1 = [800, 3000], [2, 0]
x2, y2 = [0, 3000], [1, 1]
x3, y3 = [600, 2500], [1.7, 0]


plt.plot(x1, y1,'g--', label="IS")
plt.plot(x3, y3,'o--', label="IS'")
plt.plot(x2, y2,'r-o', label="LM")
plt.ylabel('Taxa de juros', fontsize= 12)
plt.xlabel('Produção', fontsize= 12)
plt.title('Curva IS-LM: Deslocamento da curva IS', fontsize = 16)
plt.legend()

plt.show()
```

**Relação IS:** Y = C(Y - T) + I(Y,i - π² + x) + G
\
**Relação LM:** i = î

A relação IS *(investiments and savings)* mostra o que está acontecendo no mercado de bens e serviços. \
A relação LM *(liquidity and money)* and  mostra o que está acontecendo no mercado monetário.

O aumento no **nível de risco** aumenta o juros.

Dado que existe relação negativa entre a taxa de juros e o nível de atividade econômica, um aumento no nível de risco da economia implicaria em **diminuição da atividade econômica**. Isto é, o prêmio de risco causaria um aumento no juros, que é o "preço" do dinheiro. Ficaria mais difícil de conseguir crédito em geral, para que ele possa ser utilizado de forma a gerar empregos e produzir riqueza para a sociedade. \

Desde então, tem sido constante o registro de mais vendas do que compras de títulos públicos, inclusive com alguns meses fechando em rendimento nominal negativo. Na época, o Tesouro Nacional minimizou o assunto, já que o **custo da dívida continuava a ser o menor da história.**

### 2ª questão

Explique como a política monetária pode ser usada para reduzir os
efeitos mostrados acima.

#### Resposta

```python
### plotando gráfico IS-LM

x1, y1 = [800, 3000], [2, 0]
x2, y2 = [0, 3000], [1, 1]
x3, y3 = [600, 2500], [1.7, 0]
x4, y4 = [0, 3000], [0.5, 0.5]


plt.plot(x1, y1,'g--', label="IS")
plt.plot(x3, y3,'o--', label="IS'")
plt.plot(x2, y2,'r-o', label="LM")
plt.plot(x4, y4,'b-o', label="LM'" )
plt.ylabel('Taxa de juros', fontsize= 12)
plt.xlabel('Produção', fontsize= 12)
plt.title('Curva IS-LM: Deslocamento da curva IS', fontsize = 16)
plt.legend()

plt.show()
```

Seja **r** a taxa básica de juros e **x** o prêmio de risco:

*Em equilíbrio:* 

r = 2% \
x = 1% 

r + x = 3%


*Após aumento do nível de risco:* 

r = 2% \
x = 3% 

r + x = 5%

Os instrumentos que podem ser utilizados para combater a diminuição do nível de atividade econômica são, tradicionalmente, o aumento de gastos do governo, diminuição dos tributos e diminuição da taxa nominal de juros. 

Já que o problema é justamente o aumento do juros devido ao maior nível de risco, e consequente aumento do prêmio de risco, a resposta mais adequada poderia ser a diminuição da taxa básica de juros de forma que o nível de atividade/produção retorno ao patamar de equilíbrio. 

*Readequando taxas:*

**r = 0%** \
x = 3%

**r + x = 3%**

De certa forma, essa abordagem pode ser ilustrada pela ação dos bancos centrais ao redor do mundo, que, no geral, estão atualmente com taxas básicas de juros nas suas **mínimas históricas** devido ao grande nível de risco causado pela pandemia de covid-19.

### 3ª questão

Monte um gráfico no R ou Python mostrando a evolução da taxa
nominal de juros, inflação, taxa real de juros e SELIC. Apresente o seu script e comente o resultado encontrado (opcional).

```python
### criando lista para os indicadores que usaremos

series = {'Taxa média de juros' : 20714, 'Taxa SELIC' : 4189,
          'IPCA em 12 meses' : 13522}
```

```python
### criando função para obter dados do Bacen

def bcb(codigo) :
    link = f'http://api.bcb.gov.br/dados/serie/bcdata.sgs.{codigo}/dados?formato=json'
    df = pd.read_json(link)
    df.index = df['data']
    df = df['valor']
    return df
```

```python
### rodando função para cada indicador na lista series

series_df = pd.DataFrame()

for i in series.keys():
    ind = series.get(i)
    series_df[i] = bcb(ind)
    
series_df
```

```python
### criando coluna de juros real baseado nos dados da taxa SELIC e IPCA


series_df['Juros real'] = ((1 + series_df['Taxa SELIC']) / (1 + series_df['IPCA em 12 meses'])) - 1

series_df
```

```python
### plotando gráfico

series_df.plot()
plt.title('Indicadores')
plt.ylabel('Valores em %', fontsize = 12)
plt.xlabel('De 2011 até fevereiro de 2021', fontsize = 12)
plt.xticks(rotation=25)
plt.show()
```

#### Resposta

Primeiramente, é fácil perceber um significativo descolamento entre a taxa média de juros praticada em operações de crédito e a taxa básica de juros da economia, a taxa SELIC. Isso se deve, entre outros motivos, ao prêmio de risco cobrado por instituições financeiras para ofertar crédito, afinal, é muito mais provável que uma pessoa ou empresa deixe de cumprir com suas obrigações do que o governo. Ou, melhor dizendo, é muito mais difícil um governo quebrar, até porque um dos mecanismos que podem ser usados para seu financiamento é a impressão de moeda.
\
Um segundo fato notório é o fim de um ciclo de repetidas baixas na taxa SELIC, após pico em 2016. Recentemente, o COPOM decidiu aumentar a taxa em 0,75 p.p. Isso se deve principalmente a aceleração do IPCA, como é possível perceber no gráfico acima. Somado a esses fatos, é curioso notar que, atualmente, temos juros real negativo (ver gráfico abaixo). Títulos públicos como a LFT (Tesouro SELIC), estão obtendo rendimentos inferiores ao IPCA. 

```python
### plotando gráfico do juros real

series_df['Juros real'].plot()
plt.title('Juros real')
plt.ylabel('Valores em %', fontsize = 12)
plt.xlabel('De 2011 até fevereiro de 2021', fontsize = 12)
plt.xticks(rotation=25)
plt.style.use('fivethirtyeight')
plt.show()
```
