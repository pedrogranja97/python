```python
import pandas as pd
import pandas_datareader as web
import numpy as np
import matplotlib.pyplot as plt
```

##### Cotações de cada ativo

```python
ativo1 = 'LUPA3.SA'
ativo2 = 'GGBR4.SA'
ativo3 = 'SULA11.SA'
ativo4 = 'IGTA3.SA'
ativo5 = 'USIM5.SA'
ativo6 = 'EQTL3.SA'
ativo7 = 'MDIA3.SA'
ativo8 = 'CSNA3.SA'
ativo9 = 'BIDI11.SA'
```

```python
tickers = [ativo1, ativo2, ativo3, ativo4, ativo5, ativo6, ativo7, ativo8, ativo9]

carteira = web.get_data_yahoo(tickers, start = '2021-03-01', end = '2021-03-31')['Adj Close']

carteira
```

##### Rendimento de cada ativo.

```python
carteira_rend = carteira / carteira.iloc[0]
carteira_rend
```

### CDI

```python
data_inicio = '2021-03-01'
data_fim = '2021-03-31'
```

```python
def consulta_bc(codigo_bcb):
    url = 'http://api.bcb.gov.br/dados/serie/bcdata.sgs.{}/dados?formato=json'.format(codigo_bcb)
    df = pd.read_json(url)
    df['data'] = pd.to_datetime(df['data'], dayfirst=True)
    df.set_index('data', inplace=True)
    return df
```

```python
def cdi_acumulado(data_inicio, data_fim):
    cdi = consulta_bc(12)
    cdi_acumulado = (1 + cdi[data_inicio : data_fim] / 100).cumprod()
    cdi_acumulado.iloc[0] = 1
    return cdi_acumulado
```

```python
cdi = cdi_acumulado(data_inicio=data_inicio, data_fim=data_fim)
```

#### Consolidando dataframe

```python
carteira_rend['CDI'] = cdi
carteira_rend['PL'] = 100000 * (carteira_rend[ativo1] + carteira_rend[ativo2] + carteira_rend[ativo3] + carteira_rend[ativo4] + 
                               carteira_rend[ativo5] + carteira_rend[ativo6] + carteira_rend[ativo7] + carteira_rend[ativo8] + 
                               carteira_rend[ativo9] + carteira_rend['CDI'])

carteira_rend['Rend. Acumulado (%)'] = ((carteira_rend['PL'] / carteira_rend['PL'].iloc[0]) - 1) * 100 
carteira_rend['Variação diária (%)'] = carteira_rend['PL'].pct_change() * 100
carteira_rend['Variação diária (%)'].iloc[0] = 0 

carteira_rend
```

*****

   ### Letra A

```python
retorno_mensal = ((carteira_rend['PL'].iloc[-1] / carteira_rend['PL'].iloc[0]) - 1) * 100 
print('O retorno mensal foi de ' + str(round(retorno_mensal, 2)) + '%.')
```

### Letra B

```python
ibov = web.get_data_yahoo('^BVSP', start = '2021-03-01', end = '2021-03-31')['Adj Close']

Ibovespa = ((ibov / ibov.iloc[0]) - 1) * 100

print('O alpha foi de ' + str(round(retorno_mensal - Ibovespa.iloc[-1], 2)) + '%.')
```

### Letra C

```python
comparacao = carteira_rend['Rend. Acumulado (%)']

comparacao = pd.DataFrame(comparacao)

comparacao['Ibovespa'] = Ibovespa

comparacao = comparacao.rename(columns={'Rend. Acumulado (%)': 'Carteira'})

comparacao
```

```python
comparacao.plot()
plt.ylabel('Rendimento acumulado (%)', fontname = 'Franklin Gothic Book', fontsize = 12)
plt.title('Performance', fontname = 'Franklin Gothic Book', fontsize = 18)
plt.xlabel('', fontname = 'Franklin Gothic Book', fontsize = 12)
plt.ylim(-5, 40)

plt.show()
```

### Letra E

```python
risk_free = 2.75
```

```python
port_retorno = carteira_rend['Variação diária (%)'].mean() * 252
port_retorno

print('O retorno anualizado é de ' + str(round(port_retorno, 2)) + '%.')
```

```python
vol = carteira_rend['Variação diária (%)'].std() * np.sqrt(252)

print('A volatilidade anualizada da carteira é de ' + str(round(vol, 2)) + '%.')
```

```python
sharpe = (port_retorno - risk_free) / vol

print('O Índice de Sharpe da carteira é de ' + str(round(sharpe, 2)) + '.')
```
